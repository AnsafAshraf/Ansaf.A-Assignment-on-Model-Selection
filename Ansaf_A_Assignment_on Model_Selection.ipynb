# 1.KNN Model

```python
x=df1.drop('Survived',axis=1)
y=df1['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
metric_knn=[]
n_neighbors=np.arange(3,15)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score
```

```python
for k in n_neighbors:
 knn_clf=KNeighborsClassifier(n_neighbors=k)
 model_knn=knn_clf.fit(x_train,y_train)
 y_pred_knn=model_knn.predict(x_test)
 acc=accuracy_score(y_test,y_pred_knn)
 metric_knn.append(acc)
```

```python
metric_knn
```

```python
plt.plot(n_neighbors,metric_knn,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
knn_clf=KNeighborsClassifier(n_neighbors=6)
model_knn=knn_clf.fit(x_train,y_train)
y_pred_knn=model_knn.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_knn))
```

```python
print(accuracy_score(y_test,y_pred_knn))
```

```python
print(precision_score(y_test,y_pred_knn))
```

```python
print(recall_score(y_test,y_pred_knn))
```

```python
print(f1_score(y_test,y_pred_knn))
```

# a) K Fold Cross Validation for KNN Model

```python
from sklearn.model_selection import KFold
```

```python
kfold_validator_knn=KFold(10,shuffle=True,random_state=42)
```

```python
for train_index,test_index in kfold_validator_knn.split(x,y):
  print('Training Index',train_index)
  print('Testing Index',test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result_knn=cross_val_score(model_knn,x,y,cv=kfold_validator_knn)
```

```python
cv_result_knn
```

```python
np.mean(cv_result_knn)
```

# b)Stratified K Fold Cross Validation of KNN Model

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in skfold_validator.split(x,y):
  print('Training Index',train_index)
  print('Testing Index',test_index)
```

```python
skcv_result_knn=cross_val_score(model_knn,x,y,cv=skfold_validator)
```

```python
skcv_result_knn
```

```python
np.mean(skcv_result_knn)
```
